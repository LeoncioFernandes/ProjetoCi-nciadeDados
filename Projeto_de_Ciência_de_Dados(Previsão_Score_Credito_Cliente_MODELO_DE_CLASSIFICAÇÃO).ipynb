```python
#Importar o pandas e o LabelEncoder do sklearn
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### Atribuindo a base de Dados ao DataFrame

```python
#Atribuindo a base clientes.csv à variável df com o pandas
df = pd.read_csv("clientes.csv")
```

### Verificando as informações da que contém na coluna com as variáveis dependentes

### Transformando as colunas não numéricas

```python
#Transformando a coluna score_credito (que é uma String) em valores numéricos com o LabelEncoder

codificador = LabelEncoder()

df["score_credito"] = codificador.fit_transform(df["score_credito"])
```

```python
#Imprimindo novamente a coluna score_credito mas agora com seus valores transformados em números
print(df["score_credito"].value_counts())
```

```python
#Transformando as demais colunas (que são Strings) em valores numéricos com o LabelEncoder

df["score_credito"] = codificador.fit_transform(df["score_credito"])

df["profissao"] = codificador.fit_transform(df["profissao"])

df["mix_credito"] = codificador.fit_transform(df["mix_credito"])

df["comportamento_pagamento"] = codificador.fit_transform(df["comportamento_pagamento"])
```

### Dividindo a base em treino e teste

```python
#Importando o train_test_split do sklearn
from sklearn.model_selection import train_test_split

#Importando o RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

#Importando o KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Em x, atribuindo TODAS as colunas da Base, excluindo id_cliente e score_credito (que é a coluna que contém as variáveis de dependência)
x = df.drop(["id_cliente", "score_credito"], axis = 1)

#Em y, atribuindo a coluna score_credito, que contém as variáveis de dependência
y = df["score_credito"]
```

```python
#Atribução das variáveis de TREINO e TESTE, x e y
#random_state (semente de aleatoriedade fixa - gera modelos de resultados iguais à partir da mesma semente)
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
#Atribuindo o RandomForestClassifier à variável modelo_arvoredecisao
modelo_arvoredecisao = RandomForestClassifier()

#Atribuindo o KNeighborClassifier à variável modelo_knn
modelo_knn = KNeighborsClassifier()
```

```python
#Atribuindo as variáveis de Treino aos dois Modelos
#Treinando nos dois modelos

modelo_arvoredecisao.fit(x_treino, y_treino)

modelo_knn.fit(x_treino, y_treino)
```

### Escolhendo o MELHOR modelo pelo o accuracy_score

```python
#Importando o accuracy_score
from sklearn.metrics import accuracy_score
```

```python
#Calculando a acurácia (predict) com o x_teste

#Modelo de Arvore de Decisão
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)

#Modelo de KNN
previsao_knn = modelo_knn.predict(x_teste) #caso o knn não funcione: modelo_knn.predict(x_teste.to_numpy)
```

```python
#Imprimindo as acurácias para identificar o melhor modelo

print(f"Acurácia do modelo ÁRVORE DE DECISÃO: {accuracy_score(y_teste, previsao_arvoredecisao)}")
print(f"Acurácia do modelo KNN: {accuracy_score(y_teste, previsao_knn)}")
```
